## Задача 2
Сравниваем метрику конверсия в покупку. Размер выборки - 10000 элементов в каждой группе . Какой статистический критерий тут лучше всего подойдёт и почему ?

```python
import statsmodels.api as sm

n1 = 10000  # Общее число элементов в группе 1
conv1 = 0.015  # Конверсия в группе 1 (1.5%)

# Заданные значения для группы 2
n2 = 10000  # Общее число элементов в группе 2
conv2 = 0.022  # Конверсия в группе 2 (2,2%)

# число юзеров совершивших покупку в каждой группе
num_converted1 = int(n1 * conv1)
num_converted2 = int(n2 * conv2)

# Создаем таблицу сопряженности
table = [[num_converted1, n1 - num_converted1], [num_converted2, n2 - num_converted2]]

# Z-тест для разности долей
z_stat, p_value = sm.stats.proportions_ztest([num_converted1, num_converted2], [n1, n2])

# Результаты
print("Z-статистика:", z_stat)
print("p-значение:", p_value)

# Проверка статистической значимости с уровнем значимости alpha (например, 0.05)
alpha = 0.05
if p_value < alpha:
    print("Есть статистически значимые различия между конверсиями групп.")
else:
    print("Нет статистически значимых различий между конверсиями групп.")
```
